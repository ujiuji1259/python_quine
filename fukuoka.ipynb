```python
import re

def create_ascii(v):
    chrs = "^(~:"
    aa = v.replace(" ","0").replace("`","0").replace(".","0").replace("\"","0").replace("]","0").replace("!","0").replace("?","0").replace("_","0")
    for c in chrs:
        aa = aa.replace(c,"0")
    # 開業文字以外の文字列を0に置換
    aa = re.sub(r"[^0\n`]","1",aa)
    return aa
aa = create_ascii(aaraw)
aa_1 = create_ascii(aas[0])
```
